## 2920 음계
- https://www.acmicpc.net/problem/2920
- 설명
    - 다장조는 c d e f g a b C, 총 8개 음으로 이루어져있다. 이 문제에서 8개 음은 다음과 같이 숫자로 바꾸어 표현한다. c는 1로, d는 2로, ..., C를 8로 바꾼다.
    - 1부터 8까지 차례대로 연주한다면 ascending, 8부터 1까지 차례대로 연주한다면 descending, 둘 다 아니라면 mixed 이다.
    - 연주한 순서가 주어졌을 때, 이것이 ascending인지, descending인지, 아니면 mixed인지 판별하는 프로그램을 작성하시오.
- 입력 : 첫째 줄에 8개 숫자가 주어진다. 이 숫자는 문제 설명에서 설명한 음이며, 1부터 8까지 숫자가 한 번씩 등장한다.
- 출력 : 첫째 줄에 ascending, descending, mixed 중 하나를 출력한다.

```python
inputList = list(map(int, input().split(' ')))

ascending, descending = True, True
for i in range(1, len(inputList)):
    if inputList[i-1] < inputList[i]:
        descending = False
    elif inputList[i-1] > inputList[i]:
        ascending = False

if ascending:
    print("ascending")
elif descending:
    print("descending")
else :
    print("mixed")
```

## 2798 블랙잭
- https://www.acmicpc.net/problem/2798
- 입력 : 첫째 줄에 카드의 개수 N(3 ≤ N ≤ 100)과 M(10 ≤ M ≤ 300,000)이 주어진다. 둘째 줄에는 카드에 쓰여 있는 수가 주어지며, 이 값은 100,000을 넘지 않는다. 합이 M을 넘지 않는 카드 3장을 찾을 수 있는 경우만 입력으로 주어진다.
- 출력 : 첫째 줄에 M을 넘지 않으면서 M에 최대한 가까운 카드 3장의 합을 출력한다.

```python
from sys import stdin
N, M = list(map(int, stdin.readline().split()))
cards = list(map(int, stdin.readline().split()))

output = 0
finalout = 0
for i in range(N):
    if cards[i] <= M:
        output = cards[i]
        for j in range(i+1, N):
            if output + cards[j] <= M:
                output += cards[j]
                for k in range(j+1, N):
                    if output + cards[k] <= M and output + cards[k] > finalout:
                        finalout = output + cards[k]
                output -= cards[j]
print(finalout)
```

## 1874 스택 수열
- https://www.acmicpc.net/problem/1874
- 입력 : 첫 줄에 n (1 ≤ n ≤ 100,000)이 주어진다. 둘째 줄부터 n개의 줄에는 수열을 이루는 1이상 n이하의 정수가 하나씩 순서대로 주어진다. 물론 같은 정수가 두 번 나오는 일은 없다.
- 출력 : 입력된 수열을 만들기 위해 필요한 연산을 한 줄에 한 개씩 출력한다. push연산은 +로, pop 연산은 -로 표현하도록 한다. 불가능한 경우 NO를 출력한다.

```python
N = int(input())
stack = list()
save = list()
num = 1
flag = True
for _ in range(N):
    iNum = int(input())
    if iNum >= num:
        while iNum >= num:
            stack.append(num)
            save.append('+')
            if num > N:
                flag = False
                break
            num += 1
        pop = stack.pop(-1)
        save.append('-')
    else:
        pop = stack.pop(-1)
        save.append('-')
        if pop != iNum:
            flag = False
if flag:
    for i in range(len(save)):
        print(save[i])
else:
    print('NO')
```

## 1966 프린터 큐
- https://www.acmicpc.net/problem/1966
- 입력 : 첫 줄에 test case의 수가 주어진다. 각 test case에 대해서 문서의 수 N(100이하)와 몇 번째로 인쇄되었는지 궁금한 문서가 현재 Queue의 어떤 위치에 있는지를 알려주는 M(0이상 N미만)이 주어진다. 다음줄에 N개 문서의 중요도가 주어지는데, 중요도는 1 이상 9 이하이다. 중요도가 같은 문서가 여러 개 있을 수도 있다. 위의 예는 N=4, M=0(A문서가 궁금하다면), 중요도는 2 1 4 3이 된다.
- 출력 : 각 test case에 대해 문서가 몇 번째로 인쇄되는지 출력한다.

```python
testcases = int(input())

for _ in range(testcases):
    N, M = list(map(int,input().split(' ')))
    priority = list(map(int,input().split(' ')))
    queue = list()
    count = 1
    for index, item in enumerate(priority):
        queue.append([item, index])
    while True:
        maxValue = max(queue, key = lambda item: item[0])[0]
        while queue[0][0] != maxValue:
            queue.append(queue.pop(0))
        if queue[0][1] == M:
            print(count)
            break
        else:
            queue.pop(0)
            count += 1
```

## 5397 키로거
- https://www.acmicpc.net/problem/5397
- 입력 : 첫째 줄에 테스트 케이스의 개수가 주어진다. 각 테스트 케이스는 한줄로 이루어져 있고, 강산이가 입력한 순서대로 길이가 L인 문자열이 주어진다. (1 ≤ L의 길이 ≤ 1,000,000) 강산이가 백스페이스를 입력했다면, '-'가 주어진다. 이때 커서의 바로 앞에 글자가 존재한다면, 그 글자를 지운다. 화살표의 입력은 '<'와 '>'로 주어진다. 이때는 커서의 위치를 움직일 수 있다면, 왼쪽 또는 오른쪽으로 1만큼 움직인다. 나머지 문자는 비밀번호의 일부이다. 물론, 나중에 백스페이스를 통해서 지울 수는 있다. 만약 커서의 위치가 줄의 마지막이 아니라면, 그 문자를 입력하고, 커서는 오른쪽으로 한 칸 이동한다.
- 출력 : 각 테스트 케이스에 대해서, 강산이의 비밀번호를 출력한다. 비밀번호의 길이는 항상 0보다 크다.

```python
##시간초과
testcases = int(input())
for _ in range(testcases):
    pw1 = list(input())
    pw2 = list()    
    j = 0
    for key in pw1:
        if '<' == key:
            if j != 0:
                j -= 1
        elif '>' == key:
            if j != len(pw2):
                j += 1
        elif '-' == key:
            if j != 0:
                pw2.pop(j-1)
                j -= 1
        else:
            pw2.insert(j, key)
            j += 1
    print(''.join(pw2))
```

## 1920 수 찾기
- https://www.acmicpc.net/problem/1920
- 입력 : 첫째 줄에 자연수 $N(1≤N≤100,000)$이 주어진다. 다음 줄에는 N개의 정수 $A[1], A[2], …, A[N]$이 주어진다. 다음 줄에는 $M(1≤M≤100,000)$이 주어진다. 다음 줄에는 $M$개의 수들이 주어지는데, 이 수들이 $A$안에 존재하는지 알아내면 된다. 모든 정수의 범위는 $-2^{31}$ 보다 크거나 같고 $2^{31}$보다 작다.
- 출력 : $M$개의 줄에 답을 출력한다. 존재하면 1을, 존재하지 않으면 0을 출력한다.

```python
# 시간초과
def quickSort(data):
    if len(data) <= 1:
        return data
    pivot = data[0]
    leftpivot = [item for item in data[1:] if item < pivot]
    rightpivot = [item for item in data[1:] if pivot <= item]
    return quickSort(leftpivot) + [pivot] + quickSort(rightpivot)

def binarySearch(data, search_target, start_index = 0):
    mid_index = len(data) // 2
    if len(data) == 0:
        return -1
    elif data[mid_index] == search_target:
        return start_index + mid_index
    elif mid_index == 0:
        return -1
    elif data[mid_index] < search_target:
        return binarySearch(data[mid_index + 1:], search_target, start_index = start_index + mid_index + 1)
    else:
        return binarySearch(data[:mid_index], search_target, start_index = start_index)

N = int(input())
N_list = list(map(int, input().split(' ')))
M = int(input())
M_list = list(map(int, input().split(' ')))
N_list = quickSort(N_list)

for item in M_list:
    result = binarySearch(N_list, item)
    if result == -1:
        print(0)
    else:
        print(1)
```

- 출력 : 친구 관계가 생길 때마다, 두 사람의 친구 네트워크에 몇 명이 있는지 구하는 프로그램을 작성하시오.

```python
# 시간초과
testcases = int(input())
for _ in range(testcases):
    friends = int(input())
    nets = [set()]
    for _ in range(friends):
        f1, f2 = input().split(' ')
        f1_new = True
        f2_new = True
        f1_idx = 0
        f2_idx = 0
        for i in range(len(nets)):
            if f1 in nets[i]:
                f1_idx = i
                f1_new = False
            if f2 in nets[i]:
                f2_idx = i
                f2_new = False
        if f1_new == True and f2_new == True:
            nets.append(set())
            nets[-1].update([f1, f2])
        elif f1_new == True and f2_new == False:
            nets[f2_idx].add(f1)
        elif f1_new == False and f2_new == True:
            nets[f1_idx].add(f2)
        else:
            nets[f1_idx] = nets[f1_idx].union(nets[f2_idx])
            nets.pop(f2_idx)
        print(max([len(nets[i]) for i in range(len(nets))]))
```

## 2750 수 정렬하기
- https://www.acmicpc.net/problem/2750
- 입력 : 첫째 줄에 수의 개수 N(1 ≤ N ≤ 1,000)이 주어진다. 둘째 줄부터 N개의 줄에는 숫자가 주어진다. 이 수는 절댓값이 1,000보다 작거나 같은 정수이다. 수는 중복되지 않는다.
- 출력 : 첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한다.

```python
import sys
sys.setrecursionlimit(99999)

def qsort(data):
    if len(data) <= 1:
        return data
    leftpivot = list()
    rightpivot = list()
    pivot = data[0]
    for item in data[1:]:
        if item < pivot:
            leftpivot.append(item)
        else:
            rightpivot.append(item)
    return qsort(leftpivot) + [pivot] + qsort(rightpivot)

N = int(input())
nums = list()
for _ in range(N):
    nums.append(int(input()))
nums = qsort(nums)
print("\n".join(map(str, nums)))
```

- 출력 : 첫째 줄에 자리수를 내림차순으로 정렬한 수를 출력한다.

```python
def qsort2(data): # not recursive
    end_check = True
    afterdata = [data]
    while end_check:
        beforedata = afterdata[:]
        afterdata = []
        end_check = False
        for pivotdata in beforedata:
            if len(pivotdata) > 1:
                leftpivot = list()
                rightpivot = list()
                pivot = pivotdata[0]
                for item in pivotdata[1:]:
                    if item < pivot:
                        leftpivot.append(item)
                    else:
                        rightpivot.append(item)
                afterdata.append(leftpivot)
                afterdata.append([pivot])
                afterdata.append(rightpivot)
                end_check = True
            elif len(pivotdata) == 1:
                afterdata.append(pivotdata)
    result = list()
    for item in afterdata:
        result.append(item[0])
    return result

N = int(input())
nums = list()
while N:
    quotient, remainder = divmod(N, 10)
    N = quotient
    nums.append(remainder)
nums = reversed(qsort2(nums))
print("".join(map(str, nums)))
```

- 출력 : 첫째 줄부터 총 N개의 줄에 걸쳐 온라인 저지 회원을 나이 순, 나이가 같으면 가입한 순으로 한 줄에 한 명씩 나이와 이름을 공백으로 구분해 출력한다.

- 출력 : 첫째 줄에 문제의 정답을 출력한다.

```python
N, r, c = list(map(int, input().split(' ')))

def find_visit(N, r, c):
    if N == 1:
        if (r, c) == (0, 0):
            return 0
        elif (r, c) == (0, 1):
            return 1
        elif (r, c) == (1, 0):
            return 2
        elif (r, c) == (1, 1):
            return 3
    else:
        box_r = r // pow(2, N-1)
        box_c = c // pow(2, N-1)
        if (box_r, box_c) == (0, 0):
            return 0*pow(2, N-1)*pow(2, N-1) + find_visit(N-1, r - box_r*pow(2, N-1), c - box_c*pow(2, N-1))
        elif (box_r, box_c) == (0, 1):
            return 1*pow(2, N-1)*pow(2, N-1) + find_visit(N-1, r - box_r*pow(2, N-1), c - box_c*pow(2, N-1))
        elif (box_r, box_c) == (1, 0):
            return 2*pow(2, N-1)*pow(2, N-1) + find_visit(N-1, r - box_r*pow(2, N-1), c - box_c*pow(2, N-1))
        elif (box_r, box_c) == (1, 1):
            return 3*pow(2, N-1)*pow(2, N-1) + find_visit(N-1, r - box_r*pow(2, N-1), c - box_c*pow(2, N-1))

print(find_visit(N, r, c))
```

```python
# 29380KB 60ms
def search(doc, target):
    count = 0
    startIndex = 0
    while startIndex <= len(doc) - len(target):
        i = startIndex
        eq = True
        for j in range(len(target)):
            if target[j] != doc[i]:
                eq = False
                break
            i += 1
        if eq == True:
            count += 1
            startIndex = i
        else:
            startIndex += 1
    return count

doc = input()
target = input()
print(search(doc, target))
```

## 1939 중량제한
- https://www.acmicpc.net/problem/1939
- 입력 : 첫째 줄에 N, M(1≤M≤100,000)이 주어진다. 다음 M개의 줄에는 다리에 대한 정보를 나타내는 세 정수 A, B(1≤A, B≤N), C(1≤C≤1,000,000,000)가 주어진다. 이는 A번 섬과 B번 섬 사이에 중량제한이 C인 다리가 존재한다는 의미이다. 서로 같은 두 도시 사이에 여러 개의 다리가 있을 수도 있으며, 모든 다리는 양방향이다. 마지막 줄에는 공장이 위치해 있는 섬의 번호를 나타내는 서로 다른 두 정수가 주어진다. 공장이 있는 두 섬을 연결하는 경로는 항상 존재하는 데이터만 입력으로 주어진다.
- 출력 : 첫째 줄에 답을 출력한다.

```python
# 46712KB 620ms
def DFS_weight(graph, start, search_target, weight_target):
    visited = [False] * (N + 1)
    visited[start] = True
    willvisit = list()  # stack
    willvisit.append(start)
    
    while willvisit:
        cur = willvisit.pop()
        for brdg, weight in graph[cur].items():
            if not visited[brdg] and weight >= weight_target:
                visited[cur] = True
                if brdg == search_target:
                    return True
                willvisit.append(brdg)
    return False

N, M = list(map(int, input().split(' ')))
bridges = dict()
min_weight = 1000000000
max_weight = 1
for _ in range(M):  # Make bridges
    brdg = list(map(int, input().split(' ')))
    if brdg[0] not in bridges:
        bridges[brdg[0]] = dict()
        bridges[brdg[0]][brdg[1]] =  brdg[2]
    else:
        if brdg[1] not in bridges[brdg[0]]:
            bridges[brdg[0]][brdg[1]] =  brdg[2]
        else:
            bridges[brdg[0]][brdg[1]] =  max(brdg[2], bridges[brdg[0]][brdg[1]])
    if brdg[1] not in bridges:
        bridges[brdg[1]] = dict()
        bridges[brdg[1]][brdg[0]] =  brdg[2]
    else:
        if brdg[0] not in bridges[brdg[1]]:
            bridges[brdg[1]][brdg[0]] =  brdg[2]
        else:
            bridges[brdg[1]][brdg[0]] =  max(brdg[2], bridges[brdg[1]][brdg[0]])
    min_weight = min(min_weight, brdg[2])
    max_weight = max(max_weight, brdg[2])

start, end = list(map(int, input().split(' ')))

result_weight = min_weight
while min_weight <= max_weight: # Check weight_target
    weight_target = (max_weight + min_weight) // 2
    if DFS_weight(bridges, start, end, weight_target) == True:  # Check the route
        result_weight = weight_target
        min_weight = weight_target + 1
    else:
        max_weight = weight_target - 1

print(result_weight)
```

```python
# 29380KB 60ms
class Node:
    def __init__(self):
        #self.value = value
        self.left = None
        self.right = None

def preorder_traversal(tree, cur_node): # root->left->right
    print(cur_node, end='')
    if tree[cur_node].left != None:
        preorder_traversal(tree, tree[cur_node].left)
    if tree[cur_node].right != None:
        preorder_traversal(tree, tree[cur_node].right)

def inorder_traversal(tree, cur_node):    # left->root->right
    if tree[cur_node].left != None:
        inorder_traversal(tree, tree[cur_node].left)
    print(cur_node, end='')
    if tree[cur_node].right != None:
        inorder_traversal(tree, tree[cur_node].right)

def postorder_traversal(tree, cur_node):  # left->right->root
    if tree[cur_node].left != None:
        postorder_traversal(tree, tree[cur_node].left)
    if tree[cur_node].right != None:
        postorder_traversal(tree, tree[cur_node].right)
    print(cur_node, end='')

N = int(input())
tree = dict()
for _ in range(N):
    node, left, right = list(input().split(' '))
    tree[node] = Node()
    if left != '.':
        tree[node].left = left
    if right != '.':   
        tree[node].right = right

preorder_traversal(tree, 'A')
print()
inorder_traversal(tree, 'A')
print()
postorder_traversal(tree, 'A')
```

## 2250 트리의 높이와 너비

- 입력 : 첫째 줄에 노드의 개수를 나타내는 정수 N(1 ≤ N ≤ 10,000)이 주어진다. 다음 N개의 줄에는 각 줄마다 노드 번호와 해당 노드의 왼쪽 자식 노드와 오른쪽 자식 노드의 번호가 순서대로 주어진다. 노드들의 번호는 1부터 N까지이며, 자식이 없는 경우에는 자식 노드의 번호에 -1이 주어진다.
- 출력 : 첫째 줄에 너비가 가장 넓은 레벨과 그 레벨의 너비를 순서대로 출력한다. 너비가 가장 넓은 레벨이 두 개 이상 있을 때에는 번호가 작은 레벨을 출력한다.

```python
# 31128KB 100ms
class Node:
    def __init__(self, left, right):
        self.left = left
        self.right = right

def inorder_traversal(node_cur, level_cur):    # left->root->right
    global width_cur, level_max
    if tree[node_cur].left != None:
        inorder_traversal(tree[node_cur].left, level_cur + 1)

    if level_cur not in width_left:
        width_left[level_cur] = N
    if level_cur not in width_right:
        width_right[level_cur] = 1
    width_left[level_cur] = min(width_cur, width_left[level_cur])
    width_right[level_cur] = max(width_cur, width_right[level_cur])
    level_max = max(level_cur, level_max)
    width_cur += 1

    if tree[node_cur].right != None:
        inorder_traversal(tree[node_cur].right, level_cur + 1)

N = int(input().strip())
tree = dict()
width_right = dict()
width_left = dict()
root_node_candidates = dict()
for i in range(N):
    root_node_candidates[i + 1] = True

for _ in range(N):
    head, left, right = list(map(int, input().split()))
    if left == -1:
        left = None
    elif left in root_node_candidates:
        del root_node_candidates[left]
    if right == -1:
        right = None
    elif right in root_node_candidates:
        del root_node_candidates[right]
    tree[head] = Node(left, right)

if len(root_node_candidates) == 1:
    for key in root_node_candidates.keys():
        root_node = key
else :
    print('root_node Error')

width_cur = 1
level_max = 1
inorder_traversal(root_node, 1)

width_max = 0
width_max_level = 1
for level in range(1, level_max + 1):
    if width_right[level] - width_left[level] + 1 > width_max:
        width_max = width_right[level] - width_left[level] + 1
        width_max_level = level

print(width_max_level, width_max)
```

## 1927 최소 힙
- https://www.acmicpc.net/problem/1927
- 입력 : 첫째 줄에 연산의 개수 N(1≤N≤100,000)이 주어진다. 다음 N개의 줄에는 연산에 대한 정보를 나타내는 정수 x가 주어진다. 만약 x가 자연수라면 배열에 x라는 값을 넣는(추가하는) 연산이고, x가 0이라면 배열에서 가장 작은 값을 출력하고 그 값을 배열에서 제거하는 경우이다. 입력되는 자연수는 $2^{31}$보다 작다.
- 출력 : 입력에서 0이 주어진 회수만큼 답을 출력한다. 만약 배열이 비어 있는 경우인데 가장 작은 값을 출력하라고 한 경우에는 0을 출력하면 된다.

```python
# 32248KB 500ms
class Min_Heap:
    def __init__(self, data = None):
        if data == None:
            self.heap_array = [None]
        else:
            self.heap_array = [None, data]

    def insert(self, data):        
        self.heap_array.append(data)
        cur_idx = len(self.heap_array) - 1
        while True:
            parent_idx = cur_idx // 2
            if parent_idx == 0:
                break
            elif self.heap_array[parent_idx] > self.heap_array[cur_idx]:
                self.heap_array[parent_idx], self.heap_array[cur_idx] = self.heap_array[cur_idx], self.heap_array[parent_idx]
                cur_idx = parent_idx
            else:
                break
    
    def pop(self):
        if len(self.heap_array) <= 1:
            return 0

        return_data = self.heap_array[1]
        self.heap_array[1] = self.heap_array[-1]
        del self.heap_array[-1]
        cur_idx = 1

        while True:
            left_idx = cur_idx * 2
            right_idx = cur_idx * 2 + 1
            if left_idx >= len(self.heap_array):
                return return_data
            elif right_idx >= len(self.heap_array):
                if self.heap_array[left_idx] < self.heap_array[cur_idx]:
                    self.heap_array[left_idx], self.heap_array[cur_idx] = self.heap_array[cur_idx], self.heap_array[left_idx]
                    cur_idx = left_idx
                    continue
                else:
                    return return_data
            else:
                if self.heap_array[left_idx] > self.heap_array[right_idx]:
                    if self.heap_array[right_idx] < self.heap_array[cur_idx]:
                        self.heap_array[right_idx], self.heap_array[cur_idx] = self.heap_array[cur_idx], self.heap_array[right_idx]
                        cur_idx = right_idx
                        continue
                    else:
                        return return_data
                else:
                    if self.heap_array[left_idx] < self.heap_array[cur_idx]:
                        self.heap_array[left_idx], self.heap_array[cur_idx] = self.heap_array[cur_idx], self.heap_array[left_idx]
                        cur_idx = left_idx
                        continue
                    else:
                        return return_data


N = int(input().strip())
min_heap = Min_Heap()

for _ in range(N):
    x = int(input().strip())
    if x == 0:
        print(min_heap.pop())
    else:
        min_heap.insert(x)
```
